```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import healpy as hp

import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.metrics as metrics
from lsst.sims.maf.driver import runBundle
```

```python
# Pick a slicer
slicer = slicers.HealpixSlicer(nside=64)

# Configure some metrics
metricList = []
metricList.append(metrics.Coaddm5Metric())
metricList.append(metrics.MeanMetric(col='airmass'))
```

```python
# Set the database and query
dbAddress = 'sqlite:///enigma_1189_sqlite.db'
sqlWhere = 'filter = "r"'
```

```python
mafBundle = {'slicer':slicer, 'metricList':metricList, 'dbAddress':dbAddress,
             'sqlWhere':sqlWhere, 'outDir':'BundleExampleOut'}
```

```python
metricResults = runBundle(mafBundle)
```

The output files get made, including high-resolution pdf files, thumbnail pngs, and the metric values themselves in numpy save files (.npz). There's also a sqlite database that holds any summary stats and keeps track of which plots go with which metrics.

We can use use the plotting methods on the slicer to interactively make new plots

```python
metricResults.slicer.plotSkyMap(metricResults.metricValues[0], label='new plot', xlabel='mags', colorMax=28)
```

Since the metric values are healpixel maps, we can also juts go straight to healpy and plot them. So if you want a differnet projection from what we use (say, zoom-in with a gnomic projection), you can do that.

```python
i=0
hp.mollview(metricResults.metricValues[i], unit=metricResults.metricNames[i])
```

```python
hp.gnomview(metricResults.metricValues[i], rot=(5.5,-5,0), xsize=500, ysize=500, title=metricResults.metricNames[i])
```
