```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import healpy as hp

import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.metrics as metrics
from lsst.sims.maf.driver import runBundle
```

```python
# Pick a slicer
slicer = slicers.HealpixSlicer(nside=64)

summaryStats = [metrics.MedianMetric()]
# Configure some metrics
metricList = []
# What fraction of 60-day, r=20 mag flat transients are detected at least once?
metric = metrics.TransientMetric(riseSlope=0., declineSlope=0., transDuration=60., 
                                 peakTime=5., rPeak=20., metricName='Alert', summaryStatList=summaryStats)
metricList.append(metric)
# Now make the light curve shape a little more realistic. 
metric = metrics.TransientMetric(riseSlope=-5., declineSlope=0.5, transDuration=60., 
                                 peakTime=5., rPeak=20., metricName='Alert, shaped LC', summaryStatList=summaryStats)
metricList.append(metric)
#  Demand 2 points before tmax before counting the LC as detected
metric = metrics.TransientMetric(riseSlope=-5., declineSlope=0.5, transDuration=60., 
                                 peakTime=5., rPeak=20., nPrePeak=2, metricName='Detected on rise', summaryStatList=summaryStats)
metricList.append(metric)
```

```python
# Set the database and query
dbAddress = 'sqlite:///enigma_1189_sqlite.db'
sqlWhere = 'filter = "r"'
mafBundle = {'slicer':slicer, 'metricList':metricList, 'dbAddress':dbAddress,
             'sqlWhere':sqlWhere, 'outDir':'BundleExampleOut'}
```

```python
metricResults = runBundle(mafBundle)
```

```python
for i in metricResults.metricNames:
    print metricResults.metricNames[i], metricResults.summaryStats[i]
```
